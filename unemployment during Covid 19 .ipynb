```python
# Adding Day, Month, and Year we are changingthe format
df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year

df
```

```python
 # now we will calculate unemployment rate overtime
```

```python
# Group the data by date and calculate the mean   unemployment at each date
men_unemp_day=df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
# create a line plot
plt.figure(figsize=(10,6))# creating the sheet 
plt.plot(men_unemp_day.index,men_unemp_day.values,marker='o',linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title("Unemployment rate over time")
plt.grid(True)
plt.show()





```

```python
# Group the data by date and calculate the mean   unemployment at each date
emp_day=df.groupby('Date')['Estimated Labour Participation Rate (%)'].mean()
# create a line plot
plt.figure(figsize=(10,6))# creating the sheet 
plt.plot(emp_day.index,emp_day.values,marker='o',linestyle='-')
plt.xlabel('Date')
plt.ylabel('Estimated Labour Participation Rate (%)')
plt.title("Employment rate over time")
plt.grid(True)
plt.show()

```

```python
# Create the Sunburst plot
fig = px.sunburst(df, path=['Area', 'Region'], values='Estimated Unemployment Rate (%)', title='Sunburst Plot of Estimated Unemployment Rate by Region and Area')
fig.show()
```

```python
#  Box plot for finding the outliers in Region
```

```python
plt.figure(figsize=(25,10))
# create the box with the seprate colors for the region 
sns.boxplot(x='Region',y='Estimated Unemployment Rate (%)', data=df,palette='Set3')
# rotatiindg x axis for better visiblity
plt.xticks(rotation='vertical')
plt.title('Box Plot of Estimated Unemployment Rate by Region')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```
